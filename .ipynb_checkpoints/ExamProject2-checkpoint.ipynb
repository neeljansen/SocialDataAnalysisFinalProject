# Corona Virus and its' impact on crime in Chicago

The current corona-pandemic has had big impacts on societies all around the world. One of the places worst affected in the world is [United States of America](https://www.worldometers.info/coronavirus/country/us/) and therefore cities and companies have been shut down throughout the country and the americans are being forced to [stay at home](https://edition.cnn.com/2020/03/23/us/coronavirus-which-states-stay-at-home-order-trnd/index.html). This of course affects many things in the societies, one of which is crime rates.

As reported by [USAToday](https://eu.usatoday.com/story/news/investigations/2020/04/04/coronavirus-crime-rates-drop-and-domestic-violence-spikes/2939120001/) and [Forbes](https://www.forbes.com/sites/marleycoyne/2020/04/11/crime-rates-across-us-drop-amid-the-coronavirus-pandemic/#5e942087311e), crime rates throughout major cities of USA has dramatically dropped since the stay-at-home orders have been applied. In this study, we will visualize how crimes have decreased in the city of Chicago, Illinois and where the biggest impacts can be seen. A short outline of this web page is the following

* Analysis of the impact of Corona-virus in Illinois and more specifically Chicago and the Cook County. This included looking at the evolving pandemic in Illinois and how it affected movement of people.
* Crimes in the city of Chicago. Interactive plots to check the different types of crimes and how it currently has been compared to former years.
* An analysis of where crimes have occured in both 2019 and 2020 to get an overview of the locations of where the crimes have been comitted. This might show a different pattern, and show where people are obeying the stay-at-home orders in Chicago.
* Further analysis and forecast of the future. 


## Corona virus in Illinois, USA

As of 3th of May, 2020 a total of 1.2 million cases have been confirmed in the US. 62,000 of which are in the state of Illinois. The 7th worst county in the US is Cook County, which is the county containing most of Chicago. The current situation has a huge impace on the state of Illinois, and that is also why they chose from the 20th of March to close down the state, using the stay-at-home orders.

To get an overview of how the current situation is, we supply a map showing the situation in the different counties in Illinois. It is possible to use the mouse to hover over each county and get the general statistics of the situation in that specific county. Take a look at Cook County, where it can be seen, that the situation in Chicago is much worse than in the other counties. There are also a bit of peeks around the state. One can identify that the counties with the bigger cities are also the counties with the most cases and deaths due to the pandemic.

```python
# Packages
import pandas as pd
import numpy as np # For data handling
import matplotlib.pyplot as plt # For plotting
import datetime as dt
import urllib
import geojson
import folium
```

```python
# Load the data:
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
confirmedUS = pd.read_csv(url)

url2 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv'
deathsUS = pd.read_csv(url2)

# Rename the column Admin2 to County, since it is the name of the counties:
confirmedUS = confirmedUS.rename(columns={"Admin2": "County"})
deathsUS = deathsUS.rename(columns = {"Admin2": "County"})

illiData = confirmedUS[confirmedUS["Province_State"] == "Illinois"]
illiData = illiData.reset_index()
illiData.at[illiData[illiData["County"] == "McHenry"].index, 'County'] = 'Mchenry'
illiData.at[illiData[illiData["County"] == "McLean"].index, 'County'] = 'Mclean'
illiData.at[illiData[illiData["County"] == "De Witt"].index, 'County'] = 'Dewitt'
illiData.at[illiData[illiData["County"] == "McDonough"].index, 'County'] = 'Mcdonough'
illiData.at[illiData[illiData["County"] == "LaSalle"].index, 'County'] = 'Lasalle'
illiData.at[illiData[illiData["County"] == "DuPage"].index, 'County'] = 'Dupage'
illiData.at[illiData[illiData["County"] == "DeKalb"].index, 'County'] = 'Dekalb'


illiDataDeaths = deathsUS[deathsUS["Province_State"] == "Illinois"]
illiDataDeaths = illiDataDeaths.reset_index()
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "McHenry"].index, 'County'] = 'Mchenry'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "McLean"].index, 'County'] = 'Mclean'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "De Witt"].index, 'County'] = 'Dewitt'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "McDonough"].index, 'County'] = 'Mcdonough'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "LaSalle"].index, 'County'] = 'Lasalle'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "DuPage"].index, 'County'] = 'Dupage'
illiDataDeaths.at[illiDataDeaths[illiDataDeaths["County"] == "DeKalb"].index, 'County'] = 'Dekalb'
```

```python
url= 'https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/illinois-counties.geojson'
response = urllib.request.urlopen(url)
county_geo = geojson.loads(response.read())
date = "5/1/20"
# Get the data of the states:
state_data = illiData[["County",date, "Lat", "Long_"]]
state_data = state_data.reset_index()
# Make it logarithmic:
state_data[date] = np.log(state_data[date]+1)
 
# Initialize the map:
m = folium.Map(location=[40.000000, -89], zoom_start=7)

for i in county_geo.features:
    i.properties["Cases"] = round(np.exp(state_data[state_data["County"] == i.properties['name']][date].to_numpy()[0]))
    i.properties["Deaths"] = int(illiDataDeaths[illiDataDeaths["County"] == i.properties['name']][date].to_numpy()[0])

# Add the color for the chloropleth:
choro = folium.Choropleth(
 geo_data=county_geo,
 name='COVID-19',
 data=state_data,
 columns=['County', date],
 key_on='feature.properties.name',
# threshold_scale=[0, 100, 500, 1000, 20000],
 fill_color='YlGn',
 fill_opacity=0.6,
 line_opacity=0.2,
 highlight=True,
 legend_name='Amount of confirmed cases (logarithmic)'
).add_to(m)
folium.LayerControl().add_to(m)

style_function = "font-size: 12px; font-weight: bold"
choro.geojson.add_child(
    folium.features.GeoJsonTooltip(['name', 'Cases', 'Deaths'], style=style_function, labels=True))

m
```

As it can be seen, the situation in Illinois is dependent on where in the state one is. The more people, the more confirmed cases and deaths. Therefore, it seems quite relevant to focus on the Chicago area. Chicago is located in Cook County on the northeastern part of Illinois. Given that Chicago is the largest city in Illinois, it is also there most cases can be seen. This is visualized by the following plot, showing how the pandemic has evolved in the entire state of Illinois and how it compares to the evolving pandemic in Cook county.

```python
## Get data for the states (meaning total number of cases for each state, disregarding counties)
# First, sum the different states, if more entries of one state
tmp = illiData.groupby("Province_State").sum()

# Transpose the data frame:
StateData = tmp.T

# Remove unwanted rows:
StateData = StateData.drop(StateData.index[0:6])

# Rename the first column:
StateData = StateData.reset_index()
StateData = StateData.rename(columns={"index": "Date"})

# Now every column is a state (except for date), and every row corresponds to a date
```

```python
# Write the state:
state = "Illinois"
dat = StateData[state]
tmp = StateData["Date"]
dat2 = illiData[illiData["County"] == "Cook"]
dat2 = dat2.T
dat2 = dat2.drop(dat2.index[0:12])
#plt.plot(dat2.to_numpy())


# Make the plot for that state:
fig=plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(tmp, dat, label = "State of Illinois")
plt.plot(tmp, dat2.to_numpy(), label = "Cook County")
plt.axvline(x="3/21/20", color = "black", label="Stay-at-home order")
plt.xticks(np.arange(0, len(dat), step=6), tmp[np.arange(0, len(dat), step=6)], rotation=50);
plt.xlabel("Date");
plt.ylabel("Number of cases");
plt.title("Number of confirmed COVID-19 cases in "+ state);
plt.legend()
plt.grid();
```

It is clear from the plot, that the most affected place in Illinois is around Chicago in Cook County, having by far the largest proportion of the total number of cases. 

The effect on the daily life in Chicago is of course very high. Stay-at-home orders was issued in Illinois the [20th of March, 2020](https://www.chicago.gov/content/dam/city/depts/mayor/Press%20Room/Press%20Releases/2020/March/FAQsStayAtHomeOrder.pdf), and they have been extended through April. Since people are required to stay at home, huge effects can be seen. The time of the stay-at-home orders can be seen by the black line. It is quite clear that even though stay-at-home orders have veen applied, it was too late and the number of confirmed cases exploded. This might be due to people not following the orders or because people were already infected at that point.

### Movement of people in Cook County

To analyse whether or not people have been following the stay-at-home orders, it is very relevant to analyse how much movement of the people in Cook County has decreased after the orders were applied. To do so, a study performed by Descartes Labs can be used. The studies have measured the median of the longest movement a person has done in a day and the study can be found [here](https://www.descarteslabs.com/mobility/). 

The following plot shows the development of movement in Cook County. The black line in the plot shows when the stay-at-home orders were introduced. It should be noted, that the data is from the 1st of March, 2020 until the 27th of April, 2020.

```python
## Load the mobility data
url3 = 'https://raw.githubusercontent.com/descarteslabs/DL-COVID-19/master/DL-us-mobility-daterow.csv'
mobility = pd.read_csv(url3)
mobility = mobility.rename(columns = {"admin1": "State", "admin2": "County"})
illiMob = mobility[mobility["State"] == "Illinois"]
illiMob = illiMob.reset_index()
```

```python
# Plot at test. Let us focus on Cook County, which is where Chicago is.
test = illiMob[illiMob["County"] == "Cook County"]
dates = test["date"]
fig=plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(dates, test["m50"], color = "orange", label = "Distance travelled")
plt.axvline(x="2020-03-21", color = "black", label="Stay-at-home order")
plt.title("Distance travelled in Cook County, Illinois", size = 14)
plt.ylabel("Median of maximum distance travelled", size = 12)
plt.xlabel("Date", size = 12)
#plt.grid()
plt.legend()
plt.xticks(dates[::7], rotation = 45);
```

As it can be seen, already prior to the stay-at-home orders, movement of people in Cook County decreased rather rapidly. After the stay-at-home orders where introduced, one can see that the median of the maximum distance travelled is very low. This is therefore a strong indication that people are actually staying at home during the Corona pandemic. 

Another important thing about distance travelled is how it affects the society in general in Cook County and in Chicago. Thereby we will now look at the crimes in Chicago, since media has reported it decreasing in major cities of USA, including Chicago.

## Crime in Chicago

Chicago has a [higher crime rate than average](https://en.wikipedia.org/wiki/Crime_in_Chicago) in the US. The city has a high rate of homicides and other violent crimes, and street gangs are evolving throughout the city. Chicago is regareded as the most gang-infected city in America, and thereby crimes of homicides, narcotica and burglary is not uncommon in the area.

However, crimes in Chicago has seen [a rapid decline](https://nypost.com/2020/04/11/crime-rates-drop-as-coronavirus-spreads/) over the course of the stay-at-home period. In this section, we will focus on how the crime have declined over the period. This includes the type of crimes which have seen a decline, where the crimes occur and why that might be.

First, we will look at just the total number of crimes in Chicago during the period from 1st of January to the 18th of April for both year 2019 and 2020. 

The data is taken from the Chicago Police Department in the [Chicago City Portal](https://data.cityofchicago.org/Public-Safety/Crimes-2001-to-present/ijzp-q8t2).

```python
## Load the data
crimeDat = pd.read_csv("CrimesChicago.csv")
# link for downloaded with urllib
#crimesurl = 'https://data.cityofchicago.org/api/views/ijzp-q8t2/rows.csv?accessType=DOWNLOAD'
#crimeDat = pd.read_csv(crimesurl)
# https://data.cityofchicago.org/api/views/ijzp-q8t2/rows.csv?accessType=DOWNLOAD
```

```python
# Data 2020
dat2020 = crimeDat[crimeDat["Year"] >= 2020]
dat2020 = dat2020.reset_index()
dat2020["Date"] = pd.to_datetime(dat2020["Date"])
# Data 2019
dat2019 = crimeDat[crimeDat["Year"] == 2019]
dat2019 = dat2019.reset_index()
dat2019["Date"] = pd.to_datetime(dat2019["Date"])
# Prepare data
tmp = dat2020["Date"].max()
dat2020["DateNew"] = dat2020["Date"].dt.date
dat2019["DateNew"] = dat2019["Date"].dt.date
dat2019 = dat2019[dat2019["DateNew"] <= dt.date(year=2019,month=tmp.month,day=tmp.day)]
dat2020["DayMonth"] = dat2020.DateNew.map(lambda x: x.strftime('%m-%d'))
dat2020 = dat2020[dat2020["DayMonth"] != '02-29']
```

```python
# import datetime as dt
# dat2020["DateNew"] = dat2020["Date"].dt.date
# dat2019["DateNew"] = dat2019["Date"].dt.date
# dat2019 = dat2019[dat2019["DateNew"] <= dt.date(year=2019,month=4,day=18)]

# dat2020["DayMonth"] = dat2020.DateNew.map(lambda x: x.strftime('%m-%d'))
plotdata2020 = dat2020.groupby("DayMonth").size()
dates2020 = plotdata2020.index.to_numpy()
plotdata2020 = plotdata2020.to_numpy()

plotdata2019 = dat2019.groupby("DateNew").size()
dates2019 = plotdata2019.index.to_numpy()
plotdata2019 = plotdata2019.to_numpy()

plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(dates2020, plotdata2020, label = "Crimes 2020");
plt.plot(dates2020, plotdata2019, label = "Crimes 2019");
plt.axvline(x="03-21", color = "black", label="Stay-at-home order")
plt.legend();
plt.xticks(dates2020[::14], rotation = 45);
plt.xlabel("Date", size = 12);
plt.title("Crimes in Chicago", size = 14)
plt.grid()
plt.ylabel("Number of crimes", size = 12);
```

As it can be seen in the plot, the total number of crimes decrease a lot around the start of march. This is probably due to the increased focus in society on the corona-virus. The black line shows the time when the stay-at-home orders were applied to Illinois. The blue line shows 2020 and the orange line shows 2019. As it can be seen, the difference is rather huge in the total number of crimes.

It should also be noted, that the big dip in the end of January is due to [near cold records](https://www.weather.gov/lot/RecordColdJan2019) in the nothern Illinois area. Therefore much less crime was reported, since a snowstorm had essentially frozen the city.

Other crimes in the data set includes gambling, arson, homicide and others. All can be seen in the previous link to the Chicago City Bank. Of most interest is not of course how many crimes has happened in 2020, but rather comparing it to how it normally is. The easiest way to compare the crimes are by looking at how it was in 2019. Therefore, similar to the plot showing the total number of crimes, one can compare each specific type of crime to how it was last year.

In the following figure, it is possible to click the legend and select a specific crime. The number of crimes in 2020 and in 2019 will then show, so they are easy to compare.

Try to click on narcotics or theft. These are two types of crimes that have seen a rapid decrease in 2020 compared to 2019.

```python
import pandas as pd
from bokeh.palettes import Spectral8
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import Legend, Span
output_notebook()

p = figure(plot_height=500,plot_width = 1100, x_axis_type = "datetime")
p.title.text = 'Focus Crimes in Chicago. Year 2019 and 2020. Black line shows stay-at-home order.'
focuscrimes = ['ASSAULT','BATTERY','BURGLARY','DECEPTIVE PRACTICE','MOTOR VEHICLE THEFT','NARCOTICS','ROBBERY','THEFT', 'CRIMINAL DAMAGE']
tmp = dat2020.groupby("DateNew").size()
tmp2 = dat2020.groupby("DayMonth").size()
dates = tmp.index.to_numpy()
dates2 = tmp2.index.to_numpy()
stayathomedate = dates[dates == dt.date(year=2020,month=3,day=21)]
lines = {}
i = 0
legend_it = []

legend_it2 = []
legend_ = []
c = {}
for name in focuscrimes:
    color = Spectral8[0]
    color2 = Spectral8[1]
    data = dat2020[dat2020["Primary Type"] == name]
    data = data.groupby("DateNew").size().to_numpy()
    data2 = dat2019[dat2019["Primary Type"] == name]
    data2 = data2.groupby("DateNew").size().to_numpy()
    c = p.line(dates, data, line_width = 2, color = 'blue', alpha = 0.8, visible = False)
    c2 = p.line(dates, data2, line_width = 2, color = 'orange', alpha = 0.8, visible = False)
    test = p.line(dates[0], data[0], line_width = 2, color = 'blue', alpha = 0.8, visible = True)
    test2 = p.line(dates[0], data[0], line_width = 2, color = 'orange', alpha = 0.8, visible = True)
    i += 1
    legend_.append((name, [c, c2]))
    
stayathome = Span(location=stayathomedate[0],
                            dimension='height', line_color='black',
                            line_dash='dashed', line_width=1)
p.add_layout(stayathome)

legend_it.append(("2020", [test]))
legend_it2.append(("2019", [test2]))
legend2 = Legend(items = legend_it, location = (0, 0))
legend3 = Legend(items = legend_it2, location = (0, 0))
p.add_layout(legend2, 'right')
p.add_layout(legend3, 'right')

legend = Legend(items = legend_, location = (0,100))
p.add_layout(legend, 'left')
p.yaxis.axis_label = 'Number of crimes'
p.xaxis.axis_label = 'Date'
p.legend.click_policy = "hide"

#output_file("interactive_legend.html", title = "interactive_legend.py example")
show(p)
```

When one dives into the plot, it is also clear that some types of crime does not see much of a change. Motor vehicle theft is an example. There can be several reasons for this, but given people are ordered to stay at home, it might be easier to steal cars. Robbery and burglary has not changed a lot either as opposed to theft. This seems reaonable, since a robbery happens when people are at home or at a store and burglary is breaking and entering. Theft on the other hand usually happens when people are not at home. Thereby the theft numbers go down, since people are staying at home.

Just comparing the number of crimes to 2019 might be unprecise, and it would therefore be relevant to also look at other past years. This is to also argue that the decrease in number of crimes is not a common pattern, where it decreases every year. The following plot shows exactly that; the total number of crimes between 1st of March and 18th of April from year 2010 until year 2020. Again, one can choose the crime in the legend to be shown, and then see a comparisson between the years. 

```python
bardat = crimeDat[crimeDat.Year >= 2010]
```

```python
years = [i for i in range(2010,2021)]
A = np.zeros((len(years), len(focuscrimes)))
c1 = 0
for i in years:
    tmp1 = bardat[bardat["Year"] == i]
    tmp1 = tmp1.reset_index()
    c2 = 0
    for j in focuscrimes:
        tmp2 = tmp1[tmp1["Primary Type"] == j]
        tmp2 = tmp2.reset_index()
        tmp2["Date"] = pd.to_datetime(tmp2["Date"]).dt.date
        tmp2 = tmp2[tmp2["Date"] <= dt.date(year=i,month=4,day=18)]
        tmp2 = tmp2[tmp2["Date"] >= dt.date(year=i,month=3,day=1)]
        A[c1,c2] = len(tmp2.Date)
        c2 += 1
    c1 += 1
```

```python
c = 0
tmp = np.zeros(len(years), int)
for i in range(2010,2021):
    tmp[c] = str(i)
    c = c+1


A = np.c_[A, tmp]
bardataset = pd.DataFrame({focuscrimes[0]: A[:, 0], focuscrimes[1]: A[:, 1], focuscrimes[2]: A[:,2],focuscrimes[3]: A[:,3],focuscrimes[4]: A[:,4],focuscrimes[5]: A[:,5],focuscrimes[6]: A[:,6],focuscrimes[7]: A[:,7], focuscrimes[8]: A[:,8], "years": A[:,9].astype(int)})
bardataset["years"] = bardataset["years"].astype(str)
```

### Theft

First focus is on theft. Theft is a recurring crime happening throughout all of the Chicago area. It is a very common crime, which will also be evident by the maps. 

Check out the outskirts of the city. The main effect can be seen there. It is seemingly so, that the stay-at-home orders affects the outskirts of Chicago more than the downtown area. The left map shows the thefts in 2020 from 1st of March to 18th of April. The right map showws the thefts in 2019 in the same time period.

It is also quite interesting to see how the thefts are more dense in the 2019 map. This is also due to the rapid decrease in the number of thefts.

## Prediction of the future

The focus until now has been to show the reader how the crimes in Chicago has developed as a result of the COVID-19 virus and the stay-at-home orders. We saw a rapid decrease. It was shown how specific types of crimes have developed and how they were different from previous years. Further, it was shown where the crimes have happened in both 2019 and 2020 to give the user an overview of the locations of the crimes.

Now we want to show how well one can predict crimes and corona data using machine learning. It should be noted, that machine learning only takes data into account. Therefore, these predictions might not be accurate due to outer forces working on the data. However, we will show in general that predictions can be used to somewhat accurately assess data.

The goal of this section is not to understand the underlying statistics, but rather getting the user to see how statistics can perform on data like the what is presented on this webpage.

### Prediction of corona cases

The first model we have developed is an univariate convolutional neural network. We are looking at the the confirmed COVID-19 cases, and the goal of the model is to make accurate one-step predictions. We are training the model on all data available except for the last 20 days.

```python
# # corona cases
cordat = StateData[["Date",state]]
cordat["Date"] = pd.to_datetime(cordat["Date"])
```

```python
# Select focus crimes and merge data with the corona data set
#selecting focus crimes
crimegroups2020 = dat2020
crimegroups2020 = crimegroups2020.rename(columns = {"Primary Type": "Category"})
focuscrimes = ['THEFT','NARCOTICS','BATTERY','DECEPTIVE PRACTICE']
crimedat = crimegroups2020[crimegroups2020.Category.isin(focuscrimes)]

# counting icidents in each focuscrime
crimedat = crimedat.rename(columns = {"Primary Type": "Category"})
crimedat["dates"] = crimedat['Date'].dt.date
crimedat = crimedat.groupby(["dates","Category"]).count().reset_index()[["dates","Category","Description"]]
crimedat["dates"] = pd.to_datetime(crimedat["dates"])

# merging corona and crime data
table = pd.pivot_table(crimedat, values='Description', index=['dates'],
                    columns=['Category'], aggfunc=np.sum).reset_index()
mergedat =pd.merge(cordat,table, how='inner', left_on="Date", right_on="dates")
mergedat = mergedat.drop(columns=['dates'])
```

```python
n_data = np.shape(cordat)[0]
start = n_data-20
raw_seq = cordat["Illinois"].to_numpy()[0:start]
# choose a number of time steps
n_steps = 5

n_fits = 100                                                 #change n_fits
ys = cordat["Illinois"].to_numpy()
```

```python
import matplotlib.dates as mdates
yhat = np.load('yhat_uni_one_step.npy') # load   change file

plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))
# for j in range(n_fits):
#     if (j==0):
#         plt.plot(cordat.Date,yhat[:,j],linewidth=1, color='blue',label="One-step predictions") 
#     else:
#         plt.plot(cordat.Date,yhat[:,j],linewidth=1.5, color='blue') 
plt.plot(cordat.Date,ys,linewidth=2.5, color='red',label="Observed cases")
plt.plot(cordat.Date,np.mean(yhat,axis=1),linewidth=1.5, color='blue',label="One-step predictions")
plt.axvline(x=cordat.Date[start-1],color="black",linewidth=0.8,label="Training data ends", linestyle='dashed')
plt.gcf().autofmt_xdate()
plt.legend(framealpha=1, frameon=True)
plt.ylabel('Corona cases')
plt.show()
```

In the plot the blue lines are representing the one step predictions for 100 models. Convolutional neural networks have a stochastic behaviour. However, we see in the plot that the predictions are pretty accurate for predicting tomorrow. If we make our prediction for tomorrow the mean of 100 models, we would expect a good result. However, We shpuld be craeful, when using these models. The model is learning the development of the COVID-19 cases, therefore the model can only predict a development it has seen before. That is if a government initiative is introduced, the model will not be able to predict the outcome.

The example above is introduced as a proof of concept to see if convolutional neural networks can be used at all for prediction the COVID-19 cases. Now we will predict four time steps into the future. 

```python
n_data = np.shape(cordat)[0]
n_steps_in = 6
n_steps_out = 4
start = np.shape(cordat)[0]

n_fits = 100                                                    #change fits
ys = cordat["Illinois"].to_numpy()

yhat = np.load('yhat_multiple_step.npy') # load            #change file

base = pd.to_datetime(cordat["Date"].to_numpy()[n_data-1])
date_list = [base + dt.timedelta(days=x) for x in range(1,n_steps_out+1,1)]
date_list = pd.to_datetime(date_list)

plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))
plt.plot(cordat.Date,ys,linewidth=2, color='red', label='Observed cases')
plt.scatter(date_list,yhat,s=10, color='blue', label='Predictions') 
plt.gcf().autofmt_xdate()
plt.axvline(x=pd.to_datetime(cordat.Date[start-1]),color="black",linewidth=0.8,label="Training data ends", linestyle='dashed')
plt.legend(framealpha=1, frameon=True)
plt.ylabel('Corona cases')
plt.show()
```

The mean of the predictions for 100 models for the next four days can be seen. However, please take the predictions with a grain of salt.  

### Prediction of crime and corona cases

Now we will use the crime date for one step predictions. When the amount of COVID-19 cases is increasing, the amount of crime is decreasing. It can be argumented that the relationship is not multivariate, but the decreasing crime rate is merely a consequence. However, we will model the network as it was a multivariate relationship between the variables.

```python
n_data = np.shape(mergedat)[0]
n_features = np.shape(mergedat)[1]-1
# choose a number of time steps
n_steps_in = 5
n_steps_out = 1
# choose training data
start = np.shape(cordat)[0]-20

n_fits = 2                                                       # change fits
ys = mergedat[["BATTERY","DECEPTIVE PRACTICE","NARCOTICS","THEFT","Illinois"]].to_numpy()
plotdata_yhat = np.load('yhat_multivariate_s.npy')
```

```python
from bokeh.palettes import *

output_notebook()

p = figure(plot_height=500,plot_width = 1100, x_axis_type = "datetime")
p.title.text = 'Prediction of focus crimes.'
focuscrimes = ["BATTERY","DECEPTIVE PRACTICE","NARCOTICS","THEFT","Illinois"]
#tmp = dat2020.groupby("DateNew").size()
#tmp2 = dat2020.groupby("DayMonth").size()
dates = mergedat.Date[5:np.shape(plotdata_yhat)[0]]  # dates predictions
dates2 = mergedat.Date[5:np.shape(plotdata_yhat)[0]]                              #dates observed
stayathomedate = cordat.Date[59]
lines = {}
i = 0
legend_it = []

legend_it2 = []
legend_ = []
c = {}
colour = ["Blues","Greens","Oranges","Purples"]
for i in range(np.size(focuscrimes)-1):
    color = brewer[colour[i]][4][0]
    color2 = brewer[colour[i]][4][1]
    data = plotdata_yhat[5:np.shape(plotdata_yhat)[0],i]
    data2 = ys[5:np.shape(plotdata_yhat)[0],i]
    c = p.line(dates, data, line_width = 2, color = color, alpha = 0.8, visible = False)
    c2 = p.line(dates2, data2, line_width = 2, color = color2, alpha = 0.8, visible = False)
    test = p.line(dates[5], data[0], line_width = 2, color = gray(6)[0], alpha = 0.8, visible = True)
    test2 = p.line(dates2[5], data2[0], line_width = 2, color = gray(6)[3], alpha = 0.8, visible = True)
    i += 1
    legend_.append((focuscrimes[i-1], [c, c2]))
    
stayathome = Span(location=stayathomedate,
                            dimension='height', line_color='black',
                            line_dash='dashed', line_width=1)
p.add_layout(stayathome)

legend_it.append(("Predictions", [test]))
legend_it2.append(("Observed cases", [test2]))
legend2 = Legend(items = legend_it, location = (0, 0))
legend3 = Legend(items = legend_it2, location = (0, 0))
p.add_layout(legend2, 'right')
p.add_layout(legend3, 'right')

legend = Legend(items = legend_, location = (0,100))
p.add_layout(legend, 'left')
p.yaxis.axis_label = 'Number of crimes'
p.xaxis.axis_label = 'Date'
p.legend.click_policy = "hide"

#output_file("interactive_legend.html", title = "interactive_legend.py example")
show(p)
```

It is seen that the one step prediction become inaccurate after the stay at home order is introduced. However, the inaccuracy is larger after the training data set is ending and the models are treading unkown ground. Furthermore, it is seen that the amount of crime before the stay at home order follow stationary processess, this stops just days before the stay at home order.

Disregards the missing predictions at the begining, this is just a design choice as we are not using padding. We need 5 observation before we can start predicting.

Overall, it is seen that the predictions for the different types of crimes are not good. However, it is interesting to see if the COVID-19 one step predictions are improved by using the information from the crime data set.

```python
cases = list(["BATTERY","DECEPTIVE PRACTICE","NARCOTICS","THEFT","Corona cases"])
i=4
plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=20))
plt.plot(mergedat.Date,ys[:,i],linewidth=2.5, color='red',label="Observed cases")
plt.plot(mergedat.Date,plotdata_yhat[:,i],linewidth=1.5, color='blue',label="One-step predictions") 
plt.axvline(x=pd.to_datetime(cordat.Date[start-1]),color="black",linewidth=0.8,label="Training data ends", linestyle='dashed')
plt.gcf().autofmt_xdate()
plt.legend(framealpha=1, frameon=True)
plt.ylabel(cases[i])
plt.show()
```

In the plot, it is hard to distinguish the predicted and observed values. Actually, the lines a plotted almost perfectly on top of eachother. This means that even though the predictions of the different types of crimes was mediocre the predictions for COVID-19 are good. However, before we get to excited, we should again note that the model has a big limitation. The model can only predict developments it has seen before. 

Further work could be to develop a model that predicts several time steps into the future using the crime data.
